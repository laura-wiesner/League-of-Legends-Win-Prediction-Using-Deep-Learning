```python
# Create train/test data sets
from sklearn.tree import export_graphviz # needed to visualize trained decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier

X = t1.drop(columns=['win'])
y = t1['win']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train = X_train.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)
```

```python
## Let's train models over a range of depths, and score them with the test set
from sklearn.metrics import f1_score

acc_scores_dt = []
f1_scores_dt = []
max_max = 20
max_depth = np.arange(1, max_max)

for depth in max_depth:
    # Build model and train
    titanic_tree = DecisionTreeClassifier(max_depth=depth, random_state=0)
    titanic_tree.fit(X_train, y_train)

    # Test
    acc_scores_dt.append(titanic_tree.score(X_test, y_test))
    y_test_hat = titanic_tree.predict(X_test)
    f1_scores_dt.append(f1_score(y_test, y_test_hat))

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(max_depth, acc_scores_dt, 'ro-', label='Accuracy')
plt.plot(max_depth, f1_scores_dt, 'bo-', label='F1')
plt.legend()

# Print values for best test score
ix_best = np.argmax(acc_scores_dt)
print('Best accuracy score is %0.3f, for max_depth=%d' % (acc_scores_dt[ix_best], max_depth[ix_best]))
ix_best = np.argmax(f1_scores_dt)
print('Best F1-score is %0.3f, for max_depth=%d' % (f1_scores_dt[ix_best], max_depth[ix_best]))
```

```python
## We'll train models over a range of tree depths

from sklearn.ensemble import RandomForestClassifier

n_estimators = 100
max_depth = np.arange(1, max_max)
acc_scores_rf = []
f1_scores_rf = []
oob_scores_rf = []
for depth in max_depth:
    # Build model and train
    titanic_forest = RandomForestClassifier(n_estimators=n_estimators,
                                            max_depth=depth,
                                            oob_score=True,
                                            random_state=0)
    titanic_forest.fit(X_train, y_train)
    
    # Save out-of-bag (OOB) score
    oob_scores_rf.append(titanic_forest.oob_score_) 
    
    # Test
    acc_scores_rf.append(titanic_forest.score(X_test, y_test))
    y_test_hat = titanic_forest.predict(X_test)
    f1_scores_rf.append(f1_score(y_test, y_test_hat))
    
# Print values for best test score
ix_best = np.argmax(acc_scores_rf)
print('Best accuracy score is %0.3f, for max_depth=%d' % (acc_scores_rf[ix_best], max_depth[ix_best]))
ix_best = np.argmax(f1_scores_rf)
print('Best F1-score is %0.3f, for max_depth=%d' % (f1_scores_rf[ix_best], max_depth[ix_best]))
ix_best = np.argmax(oob_scores_rf)
print('Best OOB score is %0.3f, for max_depth=%d' % (oob_scores_rf[ix_best], max_depth[ix_best]))
```

```python
# Plot results
plt.figure(figsize=(10, 5))
plt.plot(max_depth, acc_scores_rf, 'ro-', label='RF: Accuracy')
plt.plot(max_depth, f1_scores_rf, 'bo-', label='RF: F1')
plt.plot(max_depth, oob_scores_rf, 'go-', label='RF: OOB')
plt.plot(max_depth, acc_scores_dt, 'ro--', alpha=0.3, label='DT: Accuracy')
plt.plot(max_depth, f1_scores_dt, 'bo--', alpha=0.3, label='DT: F1')
plt.xlabel('Maximum tree depth')
plt.ylabel('Score')
_ = plt.legend()
```

```python
from matplotlib.pyplot import figure

figure(figsize=(40, 40), dpi=80)

ix_best = np.argmax(acc_scores_rf)
max_depth_best = max_depth[ix_best]

# Build model and train
titanic_forest = RandomForestClassifier(n_estimators=n_estimators, random_state=0, max_depth=max_depth_best)
titanic_forest.fit(X_train, y_train)

plt.figure(figsize=(10,5))
column_names = X_train.columns
x_tick = np.array(range(len(column_names)))
plt.bar(x_tick, titanic_forest.feature_importances_)
plt.xticks(x_tick, column_names, rotation=90, fontsize=18)
_ = plt.ylabel('Importance', fontsize=18)
plt.xticks(fontsize = 9)
plt.title(' Full Feature Importance for High Elo Games')
plt.savefig('full_feature_high_elo.png', bbox_inches='tight')
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(0)
tf.random.set_seed(0)
```

```python
verbose = 0


X = t1.drop(columns=['win'])
y = t1['win']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train = X_train.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)

# We'll normalize our data, which often results in better models
transformer = Normalizer().fit(X_train)
X_train = transformer.transform(X_train)
X_test = transformer.transform(X_test)

# Get the number of target classes (which will be 3), and samples (150)
num_classes = len(np.unique(y))
num_samples = len(y)

```

```python
## First we construct our network architecture...

num_hidden = 5 # number of hidden neurons we'll use, in each hidden layer

# We specify the use of a sequential model--that is, a layered, feed-forward model.
model = tf.keras.Sequential()

# Start with a layer of num_hidden neurons, that take in the input features
model.add(layers.Dense(num_hidden, activation='relu'))

# Add a second layer of num_hidden neurons, fully connected to the previous layer
model.add(layers.Dense(num_hidden, activation='relu'))

# Add an output layer, with number of output units equal to our number of classes (irises).
# Again, these neurons are fully connected to those of the previous layers
model.add(layers.Dense(num_classes))


## Now we 'compile' the model. This will make it execute more quickly during training and testing.
#
# - The *optimizer* specifies what our step size and "momentum" factors are, for gradient descent
# - The *loss* function is "cross-entropy", rather than RMSE or other losses we've used
#   in the past. It's very effective when training classification models.
# - The *metrics* argument just lets TensorFlow that we want to record/monitor the model's accuracy
#   scores during training (we might have specified something else, like F1-score).
#
model.compile(optimizer=tf.keras.optimizers.Adam(0.01),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])


## Finally, we fit the model to the data. For each gradient descent iteration we'll use
# a (mini-) batch size of 30. Commonly used batch sizes range from 8 to 32.
# An "epoch" is the number of iterations that is needed such that all the training samples will
# have been used one time in training--thus, 150/30==5, in this case. We train for a lot of
# epochs, to see if the model gets to a good fit, and then starts to overfit.
epochs = 300
batch_size = num_samples//5
history = model.fit(X_train, y_train, epochs=epochs,
                    batch_size=batch_size, validation_data=(X_test, y_test))
```

```python
## We can plot the accuracy scores (for both training and test sets) that were recorded during training, 
#  as well as the cross-entropy losses...

plt.figure(figsize=(15, 5))
fs = 16 # fontsize

plt.subplot(1,2,1)
plt.plot(np.arange(1,epochs+1), history.history['loss'], label='training set loss')
plt.plot(np.arange(1,epochs+1), history.history['val_loss'], label='test set loss')
plt.legend(fontsize=fs)
plt.xlabel('Epoch', fontsize=fs)
plt.ylabel('Losses', fontsize=fs)
plt.grid(True)
plt.title('High Elo Model Loss')
plt.savefig('high_elo_loss_nn.png')

plt.subplot(1,2,2)
plt.plot(np.arange(1,epochs+1), history.history['accuracy'], label='training set acc')
plt.plot(np.arange(1,epochs+1), history.history['val_accuracy'], label='test set acc')
plt.legend(fontsize=fs)
plt.xlabel('Epoch', fontsize=fs)
plt.ylabel('Accuracies', fontsize=fs)
plt.grid(True)
plt.title('High Elo Model Accuracy')
plt.savefig('high_elo_acc_nn.png')
```
